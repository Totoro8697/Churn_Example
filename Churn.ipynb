### - First of all we need to find "churn status". If one customer made enter more than 5 days or equal to 5 days then "churn status" equal to 1, else 0.


```python
import numpy as np

churn = df.groupby("userid")["day"].agg("nunique").copy()
churn = pd.DataFrame(churn)

x =  []

for i in churn.day.values:
    if i >= 5:
        x.append(1)
    else:
        x.append(0)
        

churn["churn_status"] = np.array(x)
churn.reset_index(inplace=True)
churn
```

```python
# We have 1962 amounth of 1 and 209981 amont of 0.

new_df = churn.copy()

new_df.churn_status.value_counts()
```
